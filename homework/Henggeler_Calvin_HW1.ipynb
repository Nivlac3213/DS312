```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Perform a Calculate the line of best fit for the following points
points = [(1,1), (2,2), (3,2)]
df = pd.DataFrame(data=points, columns=['x', 'y'])
df.shape
```

```python
df.plot('x', 'y', kind='scatter')
```

```python
params = np.polyfit(df['x'], df['y'],1)
print(params)
```

```python
reg_line = params[0]*df['x'] + params[1]
```

```python
fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5));
ax.scatter(df['x'], df['y'], label='Raw Data', color='blue')
ax.plot(df['x'], reg_line, label='LinReg', color='red')
ax.legend()
```
